```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parameter simulasi
n_fragments = 10  # Jumlah fragmen kunci
fragment_length = 5  # Panjang setiap fragmen
key_length = n_fragments * fragment_length  # Total panjang kunci

# Fungsi untuk menghasilkan fragmen kunci acak
def generate_key_fragment(length):
    return ''.join(np.random.choice(list("0123456789ABCDEF"), size=length))

# Simulasi untuk menghasilkan kunci enkripsi
def generate_encryption_key(n_fragments, fragment_length):
    fragments = []
    for _ in range(n_fragments):
        fragment = generate_key_fragment(fragment_length)
        fragments.append(fragment)
    encryption_key = ''.join(fragments)
    return encryption_key, fragments

# Jalankan simulasi
encryption_key, fragments = generate_encryption_key(n_fragments, fragment_length)

# Membuat tabel dengan pandas
data = {
    "Fragmen Ke": list(range(1, n_fragments + 1)),
    "Fragmen Kunci": fragments
}
df_fragments = pd.DataFrame(data)

# Menampilkan hasil simulasi
print("Hasil Simulasi Penentuan Kunci Enkripsi:")
print(df_fragments)
print("\nKunci Enkripsi yang Dihasilkan:")
print(encryption_key)

# Visualisasi distribusi panjang fragmen
plt.bar(data["Fragmen Ke"], [len(fragment) for fragment in fragments], color='skyblue', edgecolor='black')
plt.title("Distribusi Panjang Fragmen Kunci")
plt.xlabel("Fragmen Ke")
plt.ylabel("Panjang Fragmen")
plt.xticks(data["Fragmen Ke"])
plt.grid(axis='y')
plt.show()
```
